```python
#--- imports for scraping data
import pandas as pd
import requests
from bs4 import BeautifulSoup
from IPython.display import HTML
import random
```

```python
brands = ['audi', 'bmw', 'mercedes-benz', 'opel', 'volkswagen', 'renault',
          'alfa-romeo', 'chevrolet', 'citroen', 'dacia', 'fiat', 'honda',
          'hyundai', 'jeep', 'kia', 'mazda', 'mitsubishi', 'nissan', 'peugeot',
          'porsche', 'seat', 'skoda', 'tesla', 'volvo', 'toyota']
```

```python
base_url = "https://www.autoscout24.com/lst/"
#-- brand name
additional_url_1 = "?atype=C&desc=0&page="
#-- page number
additional_url_2 = "&search_id=8olbd9f4k5&sort=standard&source=listpage_pagination&ustate=N%2CU"
```

```python
all_cards = []
for brand in brands:

  for i in range(1,15):
    print(f"Scraping for brand: {brand}, and page: {i}")
    url = base_url + brand + additional_url_1 + str(i) + additional_url_2
    snapshot = requests.get(url)
    print(f"status for the page: {snapshot}")
    raw_html = snapshot.text
    soup = BeautifulSoup(raw_html, 'html.parser')
    articles = []
    articles = soup.select('.cldt-summary-full-item')
    all_cards.extend(articles)
    print(f"Scraping page: {i} ended.")
    print("=======================================")
```

```python
len(all_cards)
```

```python
scraped_data = []
for i in range(1, len(all_cards)):
  article = all_cards[i]

  data_guid = article['data-guid']
  data_make = article['data-make']
  data_model = article['data-model']
  data_mileage = article['data-mileage']
  data_price = article['data-price']

  try:
      price = article.find('p', class_='Price_price__APlgs').text.strip()
  except AttributeError:
      price = None

  try:
      mileage = article.find('span', {'data-testid': 'VehicleDetails-mileage_road'}).text.strip()
  except AttributeError:
      mileage = None

  try:
      first_registration = article.find('span', {'data-testid': 'VehicleDetails-calendar'}).text.strip()
  except AttributeError:
      first_registration = None

  try:
      fuel_type = article.find('span', {'data-testid': 'VehicleDetails-gas_pump'}).text.strip()
  except AttributeError:
      fuel_type = None

  try:
      seller_name = article.find('span', {'data-testid': 'sellerinfo-company-name'}).text.strip()
  except AttributeError:
      seller_name = None

  scraped_data.append({
        "GUID": data_guid,
        "Make": data_make,
        "Model": data_model,
        "Mileage": mileage,
        "Price": price,
        "First Registration": first_registration,
        "Fuel Type": fuel_type,
        "Seller": seller_name
    })
```

```python
data = pd.DataFrame(scraped_data)
```
